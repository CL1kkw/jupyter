```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from seaborn import histplot
```

```python
df = pd.read_csv('train.csv')
```

```python
df
```

```python
sns.boxplot(y="Age",x="Sex",data=df)
```

```python
age_bins = [0,18,30,50,100]
age_label = ['0-18','19-30','31-50','51-100']
df['Ages'] = pd.cut(df['Age'],bins=age_bins, labels=age_label)
survived = df[df['Survived'] == 1]

survivors_by_class_sex = survived.groupby(['Ages', 'Sex'],as_index=False)["Survived"].sum()

plt.figure(figsize=(10, 6))
sns.barplot(x='Ages', y='Survived', hue='Sex', data=survivors_by_class_sex)
plt.title('Number of Survivors by Passenger Class and Sex')
plt.xlabel('Passenger Class')
plt.ylabel('Number of Survivors')
plt.show()
```
